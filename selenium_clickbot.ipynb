```python
driver = webdriver.Firefox()
driver.get("http://www.urban75.com/Mag/java7.html")
time.sleep(2)
elem = driver.find_element_by_xpath('/html/body/center[1]/table[2]/tbody/tr[3]/td/form/input[3]')
elem.click()
clickhere = driver.find_element_by_xpath('/html/body/center[1]/table[2]/tbody/tr[3]/td/form/input[4]')
while True:
    clickhere.click()
driver.close()
```
